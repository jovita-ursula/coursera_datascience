# Arithmetic Expressions

Arithmetic expressions in programming involve mathematical operations such as addition, subtraction, multiplication, and division.

We perform basic arithmetic operations using the `+`, `-`, `*`, `/`and `%` operators. Each operation returns a value that can be used in further expressions or assigned to a variable.


```python
# Multiply and add two numbers
num1 = 5
num2 = 7

# Multiply the numbers
result1 = num1 * num2

# Add the numbers
result2 = num1 + num2

# Print the results
print("Multiplication result:", result1)
print("Addition result:", result2)
```
